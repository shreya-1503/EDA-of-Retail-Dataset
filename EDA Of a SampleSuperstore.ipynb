# EDA

```python
# correlation between all the columns using heatmap
sns.heatmap(data.corr())
plt.show()
```

```python
sns.pairplot(data, hue = 'Region')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data, x = 'Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Total Order')
plt.title('Total Order by Ship Mode')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='Category', y='Sales', data=data)
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
sns.scatterplot(x = 'Profit', y = 'Discount', data = data )
plt.xlabel('Profit')
plt.ylabel('Discount')
plt.title('Profit vs Discount')
plt.show()
```

```python
sns.boxplot(x = 'Region', y = 'Profit', data = data)
plt.xlabel('Region')
plt.ylabel('Profit')
plt.title('Profit by Region')
plt.show()
```

```python
# creating pair plot of numerical columns
sns.pairplot(data[['Sales', 'Quantity', 'Discount', 'Profit']])
plt.show()
```

```python
plt.hist(data['Sales'], bins=20)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
sub_category_counts = data['Sub-Category'].value_counts()
sub_category_counts.plot(kind='bar')
plt.title('Quantity Sold by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
category_sales = data.groupby('Category')['Sales'].sum()
category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales by Category')
plt.ylabel('')
plt.show()
```

# Insights after Performing EDA on SampleSuperstore Dataset

1. Technology category accounted for the highest sales at 36.4% of total sales.

2. Office supplies category had the lowest sales, contributing only 31.3% to the total sales.

3. Binders were the most sold items, indicating high demand for office organization supplies.

4. Copiers were the least sold items, suggesting they may be high-value products or less commonly purchased items. 

5. Standard class shipping method had the highest number of orders, indicating it was the most preferred shipping      option among customers.

6. Same day shipping had the least number of orders, possibly due to its higher cost or limited availability.
